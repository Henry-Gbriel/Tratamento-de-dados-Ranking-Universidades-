```python
#Mudança nome colunas

df = df.rename(columns={
    'name': 'nome',
    'scores_teaching': 'pontuações_ensino',
    'scores_research': 'pontuações_pesquisa',
    'scores_citations': 'pontuações_citações',
    'scores_industry_income': 'pontuações_renda_da_indústria',
    'scores_international_outlook': 'pontuações_perspectiva_internacional',
    'record_type': 'tipo_de_registro',
    'member_level': 'nível_de_membro',
    'location': 'localização',
    'stats_number_students': 'estatísticas_número_de_estudantes',
    'stats_student_staff_ratio': 'estatísticas_proporção_aluno_equipe',
    'stats_pc_intl_students': 'estatísticas_porcentagem_estudantes_internacionais',
    'stats_female_male_ratio': 'estatísticas_proporção_feminino_masculino',
    'subjects_offered': 'disciplinas_oferecidas',
    'closed': 'fechado',
    'unaccredited': 'não_acreditado',
    'overall_score': 'pontuação_geral'})

print(df.columns)

df.to_csv('univesidade.csv', index=False)
display(df)
```

```python
#Vamos fazer as devidas limpeza
df_limpo = df.copy()

#Corrigir número de estudantes
df_limpo['estatísticas_número_de_estudantes'] = df_limpo['estatísticas_número_de_estudantes'].str.replace(',', '')

df_limpo['estatísticas_número_de_estudantes'] = pd.to_numeric(df_limpo['estatísticas_número_de_estudantes'], errors='coerce')


#Corrigir pontuação geral
df_limpo['pontuação_geral'] = pd.to_numeric(df_limpo['pontuação_geral'], errors='coerce')

#Separar proporção feminino:masculino
proporcao_split = df_limpo['estatísticas_proporção_feminino_masculino'].str.extract(r'(\d+):(\d+)')
df_limpo['proporcao_feminino'] = pd.to_numeric(proporcao_split[0], errors='coerce')
df_limpo['proporcao_masculino'] = pd.to_numeric(proporcao_split[1], errors='coerce')

df = df_limpo.copy()

df.to_csv('univesidade.csv', index=False)
display(df_limpo)
df.to_excel('universidade.xlsx', index=False)
```

```python
#Unificar os valores em uma variavel
nome = df['nome'].unique()

# Dicionário com os nomes traduzidos
traducoes = {} #utilizado como comparador

#Para traduzir todos os nomes precisamos passar ou interar linha por linha
for nome in nome:

  try:
    traducao = GoogleTranslator(source='auto', target='pt').translate(nome)
    #Não foi feito salvamento pois falta indexar o resultado final
    print(traducao)
  except Exception as e:
    print(f"Erro ao traduzir '{nome}': {e}")

#Substituir valores no dataframe
df['nome'] = df['nome'].replace(traducoes)

#excluir coluna proporções masculino e feminino
df = df.drop(columns=['estatísticas_proporção_feminino_masculino'])

#Salvar excel para processamento
df.to_excel('universidade.xlsx', index=False)
```

```python
nome = df['nome'].unique()
localizacao = df['localização'].unique()
tipo_de_registro = df['tipo_de_registro'].unique()

tradu_n = {}
tradu_l = {}
tradu_t = {}


for nome in nome:
  try:
    traducao = GoogleTranslator(source='auto', target='pt').translate(nome)
    tradu_n[nome] = traducao
  except Exception as e:
    print(f"Erro ao traduzir '{nome}': {e}")

for localizacao in localizacao:
  try:
    traducao = GoogleTranslator(source='auto', target='pt').translate(localizacao)
    tradu_l[localizacao] = traducao
  except Exception as e:
    print(f"Erro ao traduzir '{localizacao}': {e}")

for tipo_de_registro in tipo_de_registro:
  try:
    traducao = GoogleTranslator(source='auto', target='pt').translate(tipo_de_registro)
    tradu_t[tipo_de_registro] = traducao
  except Exception as e:
    print(f"Erro ao traduzir '{tipo_de_registro}': {e}")

df['nome'] = df['nome'].replace(tradu_n)
df['localização'] = df['localização'].replace(tradu_l)
df['tipo_de_registro'] = df['tipo_de_registro'].replace(tradu_t)

#excluir coluna proporções masculino e feminino
df = df.drop(columns=['estatísticas_proporção_feminino_masculino'])

df.to_csv('universidade.csv', index=False)
df.to_excel('universidade.xlsx', index=False)
print("Salvamento feito")
```

```python
df = pd.read_excel('/content/universidade.xlsx')

#Vamos tonar os nomes das colunas mais objetivos
colunas_renomeadas = {
    'nome': 'Universidade',
    'pontuações_ensino': 'Ensino',
    'pontuações_pesquisa': 'Pesquisa',
    'pontuações_citações': 'Citações',
    'pontuações_renda_da_indústria': 'Receita da Indústria',
    'pontuações_perspectiva_internacional': 'Perspectiva Internacional',
    'tipo_de_registro': 'Tipo de Registro',
    'nível_de_membro': 'Nível de Membro',
    'localização': 'País',
    'estatísticas_número_de_estudantes': 'Nº de Estudantes',
    'estatísticas_proporção_aluno_equipe': 'Aluno por Docente',
    'estatísticas_porcentagem_estudantes_internacionais': '% Estudantes Internacionais',
    'disciplinas_oferecidas': 'Disciplinas',
    'fechado': 'Fechada',
    'não_acreditado': 'Não Acreditada',
    'pontuação_geral': 'Pontuação Geral',
    'proporcao_feminino': '% Mulheres',
    'proporcao_masculino': '% Homens'
}
df.rename(columns=colunas_renomeadas, inplace=True)

#Limpar colunas com valores númericos
df['% Estudantes Internacionais'] = (
    df['% Estudantes Internacionais']
    .astype(str)
    .str.replace('%', '', regex=False)
    .replace('', None)
    .astype(float)
)

df['Nº de Estudantes'] = (
    df['Nº de Estudantes']
    .astype(str)
    .str.replace(',', '', regex=False)
    .replace('', None)
    .astype(float)
)

#Verificar total estudantes
df['% Total'] = df['% Mulheres'] + df['% Homens']


df['Alta Internacionalização'] = df['% Estudantes Internacionais'] > 30 #Considerar valor acima disso

#Criar parametros para pontuação
def faixa_pontuacao(score):
    if score >= 95:
        return 'A (95+)'
    elif score >= 85:
        return 'B (85-94)'
    elif score >= 70:
        return 'C (70-84)'
    else:
        return 'D (<70)'

df['Faixa de Pontuação'] = df['Pontuação Geral'].apply(faixa_pontuacao)
df.sort_values(by='Pontuação Geral', ascending=False, inplace=True)

def categoria_tamanho(estudantes):
    if estudantes < 5000:
        return 'Pequeno'
    elif estudantes < 15000:
        return 'Médio'
    else:
        return 'Grande'

df['Tamanho da Universidade'] = df['Nº de Estudantes'].apply(categoria_tamanho)

#Removendo colunas desnecessarias
df.drop(columns=['Tipo de Registro', 'Nível de Membro', 'Fechada', 'Não Acreditada', '% Total'], inplace=True)

#Como saber se tem todos dados primordiais? Criar um buscador com True ou False
colunas_chave = ['Ensino', 'Pesquisa', 'Citações', 'Receita da Indústria',
                 'Perspectiva Internacional', 'Nº de Estudantes',
                 '% Estudantes Internacionais', 'Pontuação Geral']
df['Dados Completos'] = df[colunas_chave].notnull().all(axis=1)


df.to_csv('universidade.csv', index=False)
df.to_excel('universidade.xlsx', index=False)
print("Concluído")
```

<H1> Estilização do arquivo EXCEL </H1>

```python
nome_arqui = 'universidade_final.xlsx'

#Arquivo sendo relacionado para estilização
arqui = openpyxl.load_workbook(nome_arqui)
currentSheet = arqui.active

#Ajuste para estizar arquivo excel
def ajuste_larg():
  for t_coluna in currentSheet.columns:
    max_compri = max(len(str(cell.value)) if cell.value else 0 for cell in t_coluna)
    ajuste_larg = (max_compri + 2) * 1.0
    currentSheet.column_dimensions[t_coluna[0].column_letter].width = ajuste_larg

def esti_cabecalho():
  header_fill = PatternFill(fill_type="solid", start_color="70AD47", end_color="70AD47") #preenchimento
  header_font = Font(color="FFFFFF", bold=True) # Fonte

  for cell in currentSheet[1]:
    cell. fill = header_fill
    cell.font = header_font
    cell.alignment = Alignment(horizontal="center")

def esti_linha():
  #Efeito de preenchimento
  fill_odd = PatternFill(fill_type="solid", start_color="E2EFDA", end_color="E2EFDA") #branca
  fill_even = PatternFill(fill_type="solid", start_color="FFFFFF", end_color="FFFFFF") #verde

  #Alternando entre as cores
  for row in range(2, currentSheet.max_row + 1):
    fill = fill_odd if row % 2 == 1 else fill_even
    for cell in currentSheet[row]:
      cell.fill = fill

ajuste_larg()
esti_cabecalho()
esti_linha()

arqui.save(nome_arqui)
print("Concluído")
```
